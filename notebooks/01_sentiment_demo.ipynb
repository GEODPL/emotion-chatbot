```python
# Sentiment Analysis Demo
Το συγκεκριμένο notebook παρουσιάζει ένα παράδειγμα ανάλυσης συναισθήματος με την αξιοποιήση ενός προεκπαιδευμένου μοντέλου (Hugging Face Transformers) στο πλαίσιο διπλωματικής εργασίας για ένα συναισθηματικά ευφυές chatbot.

Χρησιμοποιούμε το μοντέλο `distilbert-base-uncased-finetuned-sst-2-english`, σκοπός του οποίου είναι να ταξινομεί κείμενα ως **POSITIVE** ή **NEGATIVE**.

#  Εισαγωγή βιβλιοθηκών
from transformers import pipeline

#  Φόρτωση έτοιμου pipeline για ανάλυση συναισθήματος
sentiment_pipeline = pipeline("sentiment-analysis")

#  Παραδείγματα προτάσεων προς ανάλυση
texts = [
    "I'm so happy right now",
    "My heart is full of joy",
    "Everything feel just right.",
    "I'm so proud of what I've achieved!",
    "I can't stop smiling!.",
    "Today has been wonderful.",
    "I feel so lost lately.",
    "My heart feels heavy",
    "I'm completely exhausted",
    "I can't handle this anymore",
    "I feel empty inside",
    "I'm doing okay, I guess..."
]  

#  Εκτέλεση ανάλυσης συναισθήματος
for text in texts:
    result = sentiment_pipeline(text)[0]
    label = result['label']
    score = round(result['score'], 2)
    print(f"{text} → {label} (confidence: {score})")

##  Συμπεράσματα
- Οι θετικές και αρνητικές φράσεις αναγνωρίζονται σωστά.
- Οι ουδέτερες προτάσεις (π.χ. "I'm doing okay, I guess...".") έχουν χαμηλότερο σκορ και μπορεί να τοποθετούνται λάθος.
- Το μοντέλο λειτουργεί καλά για αγγλικά προς το παρόν.

###  Επόμενα βήματα:
- Θέλω να δοκιμάσω ελληνικές προτάσεις (π.χ. μεταφρασμένες ή απευθείας γραμμένες απο εμένα)
- Πρέπει να γίνει αναζήτηση και fine-tuning σε **ελληνικό dataset**
- Ενσωμάτωση σε συνομιλιακό chatbot
    


```
